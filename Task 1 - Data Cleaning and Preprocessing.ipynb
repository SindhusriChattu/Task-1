## Check unique values before standardization

```python
# Check unique values before standardization

print(df['country'].unique())
print(df['category'].unique())
print(df['device_type'].unique())
```

## Data Standardization

```python
# Data Standardization

# Remove extra spaces and standardize text format
df['country'] = df['country'].str.strip().str.title()

df['category'] = df['category'].str.strip().str.title()

df['device_type'] = df['device_type'].str.strip().str.title()
```

```python
# Check unique values after standardization

print(df['country'].unique()[:10])
print(df['category'].unique()[:10])
print(df['device_type'].unique()[:10])
```

## Convert date formats to a consistent type (e.g., dd-mm-yyyy).

```python
# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])

# Convert to dd-mm-yyyy format
df['date'] = df['date'].dt.strftime('%d-%m-%Y')
```

## Column Header Standardization

```python
# Check column names before renaming
print("Before Renaming:")
print(df.columns)

# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Check column names after renaming
print("After Renaming:")
print(df.columns)
```

## Checking data types

```python
# Check data types
print(df.dtypes)
```

```python
df.to_csv("sales_cleaned.csv", index=False)

print("Cleaned dataset saved successfully!")
```
